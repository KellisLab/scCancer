## mregDC as source

```r
#path 
path_to_target_files <- '/path_to_folder/ccc_filtered_plots_compiled/cDC_source'
```

```r
all_num_ixns_by_subtype_target <- c()

for(file in list.files(path_to_target_files)) {
    
    target_table <- read.csv(paste0(path_to_target_files, '/', file), 
        row.names = 'X')
    
    
    num_ixns_by_subtype <- target_table %>% group_by(target) %>% summarize(n = n())
    all_num_ixns_by_subtype_target <- rbind(all_num_ixns_by_subtype_target, num_ixns_by_subtype)
    
}
```

```r
ordered_subtypes <- subtypes[subtypes %in% all_num_ixns_by_subtype_target$target]
```

```r
celltype_mapping$target <- celltype_mapping$subtype
#merge dataframes based on the subtype 
all_num_ixns_by_subtype_target <- merge(all_num_ixns_by_subtype_target, celltype_mapping[c('target','reannotated')], by = 'target')

colnames(all_num_ixns_by_subtype_target) <- c('target', 'n', 'celltype')

all_num_ixns_by_subtype_target$target <- factor(all_num_ixns_by_subtype_target$target, levels = ordered_subtypes)
all_num_ixns_by_subtype_target$celltype <- factor(all_num_ixns_by_subtype_target$celltype, levels = cell_types)
```

```r
data <- data.frame(all_num_ixns_by_subtype_target)
```

```r
options(repr.plot.width= 12, repr.plot.height = 10)
ggplot(data, aes(x = celltype, y = n, fill = target)) +
  geom_col() + theme(panel.background = element_blank(), 
             axis.line = element_line(colour = "black"), 
             axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1)) + 
            scale_fill_manual(values = colors) + xlab('subtype target') +
    ylab('number of interactions') +
    ggtitle('Number of differential/exclusive mregDC interactions by subtype, 
mDC as source')

#ggsave('./ccc_mDC_source_ixns_summary_plot.pdf', width = 11, height = 6)
```
