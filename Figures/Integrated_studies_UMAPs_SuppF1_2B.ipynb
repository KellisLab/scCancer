```r
celltype_abundance_table <- objects.integrated@meta.data %>% group_by(cell_type, study) %>% summarize(n = n())
```

```r
celltype_abundance_table$cell_type <- factor(celltype_abundance_table$cell_type, 
                       levels=c('Tumor', 'CD8 T', 'CD4 T', 'Other T', 'Cycling T', 'NK', 'B', 'Monocyte/Macrophage',
                                  'cDC', 'pDC', 'Mast', 'Endothelial', 'Fibroblast', 'Keratinocyte'))
                                                            
```

```r
library(ggplot2)
```

```r
# Stacked
labels <- c("This study", "Tirosh_et_al", "Jerby-Arnon_et_al", "Sade-Feldman_et_al", "Zhang_et_al")

ggplot(celltype_abundance_table, aes(fill=cell_type, y=n, x=study)) + 
    geom_bar(position="stack", stat="identity", width = 0.6) + aes(ymin=0) + 
scale_fill_manual("legend", values = rev(c('Endothelial' = 'orange',
               'Keratinocyte' = 'violet', 'Fibroblast' = 'yellow', 
               'Mast' = 'brown', 'pDC' = 'purple3',
               'cDC' = 'mediumpurple1', 'Cycling T' = '#00F7ff', 
               'NK' = 'blue', 'B' = 'lightblue1', 
               'Other T' = '#00B050', 'Monocyte/Macrophage' = 'magenta',
               'CD4 T' = 'forestgreen', 'CD8 T' = 'skyblue3', 
               'Tumor' = 'coral2'))) + 
                theme_classic() + 
                theme(axis.text.x = element_text(angle = 45, hjust = 1, vjust= 1)) + #add theme 
                labs(x = "Study", y = "number of cells") + 
                scale_x_discrete(label = labels)
```
