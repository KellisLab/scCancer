```r
path <- '/home/cbw3/data/figure_generation/Supplementary_Tables/Supplementary_Table_2/'
all_tumor_markers <- readRDS(paste0(path, 'ST2_all_annot_tumor_markers_SCT_data_takeda_obj_w_all_tumor_mp.rds'))
st_save_to <- 'ST2_SCT_data_tumor_marker_csvs/'
```

```r
for(cluster in unique(all_tumor_markers$cluster)) {
    
    df <- all_tumor_markers[all_tumor_markers$cluster == cluster,]
    
    #save as table
    cluster.filename <- gsub('-', '', cluster)
    cluster.filename <- gsub(' ', '_', cluster.filename)
    cluster.filename <- gsub('/', '_', cluster.filename)
    
    write.csv(df, paste0('./', cluster.filename, '.csv'))
    
}
```
